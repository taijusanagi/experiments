# Liner Regression

Logistic regression models the probability of a binary outcome using a sigmoid function. It outputs values between 0 and 1 and is used for classification.

```python
from sklearn.datasets import make_classification
from sklearn.linear_model import LogisticRegression
import matplotlib.pyplot as plt
from mlxtend.plotting import plot_decision_regions

X, y = make_classification(n_samples=200, n_features=2, n_redundant=0, n_classes=2, random_state=1)
model = LogisticRegression()
model.fit(X, y)

plot_decision_regions(X, y, clf=model)
plt.title("Logistic Regression Decision Boundary")
plt.show()
```
