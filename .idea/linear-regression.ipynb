# Liner Regression

Linear regression finds a straight line that best fits the data. It models the relationship between a dependent variable and one or more independent variables.

```python
from sklearn.linear_model import LinearRegression
import numpy as np
import matplotlib.pyplot as plt

X = np.random.rand(100, 1) * 10
y = 3 * X.flatten() + 4 + np.random.randn(100) * 2

model = LinearRegression()
model.fit(X, y)

# Plot
plt.scatter(X, y, color='blue')
plt.plot(X, model.predict(X), color='red')
plt.title("Linear Regression Fit")
plt.xlabel("X")
plt.ylabel("y")
plt.show()
```
